### Data Visulization


```python
sns.set_style('whitegrid')
sns.countplot( x="gender" , data = exam , color="y")
```

```python

plt.hist(exam["total score"], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Total Score')
plt.ylabel('Count Of Students')
plt.show()
```

```python
sns.pairplot(exam)
```

```python
sns.barplot(data= exam , x="race/ethnicity" , y="total score" , palette ="Spectral")
```

```python
plt.figure(figsize=(16,8))
exam.groupby('parental level of education')['gender'].nunique().plot(kind='bar', color='m')
plt.title('Team played in each season', fontsize=15)
plt.xlabel('Parents Education', fontsize=15)
plt.xticks(rotation=0,fontsize=15)
plt.ylabel('lunch', fontsize=15)
```
